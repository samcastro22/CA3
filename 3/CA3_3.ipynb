# Part 3. Dictionaries and Sets

12.- Use the marco geoestadístico (https://www.inegi.org.mx/app/biblioteca/ficha.html?upc=889463807469) 2010 
and 2020 (https://www.inegi.org.mx/app/biblioteca/ficha.html?upc=702825292812) to 
obtain the “Áreas Geoestadísticas Básicas” (AGEBs) from Mérida, Yucatán. AGEBs 
evolve on time. Use sets to find: 

* AGEBs which remains from 2010 on 2020. 

* New AGEBs on 2020. 

* AGEBs that disappear from 2010 to 2020. 

Make a map with the following visualization to observe all three cases.  

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
Mex_2010 = gpd.read_file('AGEB_urb_2010_5.dbf')
Mex_2010.head()
```

```python
# Filtering by "Clave geoestadistica concatenada"
Mex_merida_2010 = Mex_2010[Mex_2010['CVEGEO'].str.startswith('31050')]
Mex_merida_2010
```

```python
Mex_2020 = gpd.read_file('31a.dbf') 
Mex_2020
```

```python
# Filtering by "Clave de Municipio" Merida, Yucatan (050)
Mex_merida_2020 = Mex_2020[Mex_2020['CVE_MUN'] == "050"]
Mex_merida_2020
```

```python
# The only feature they share is the "CVEGEO" which will be our key now that we will be working with sets
set_2010 = set(Mex_merida_2010["CVEGEO"])
set_2020 = set(Mex_merida_2020["CVEGEO"])
```

### AGEBs which remains from 2010 on 2020.

```python
agebs_remain = set_2010.intersection(set_2020)
print("AGEBs that remain from 2010 to 2020:")
agebs_remain
```

```python
filtered_agebs_2020 = Mex_merida_2020[Mex_merida_2020['CVEGEO'].isin(agebs_remain)]

fig, ax = plt.subplots(figsize=(15, 15))
filtered_agebs_2020.plot(ax=ax, color='purple', edgecolor='black')
plt.title('AGEBs that Remain from 2010 to 2020')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

### New AGEBs on 2020.

```python
new_agebs_2020 = set_2020.difference(set_2010)
print("\nNew AGEBs 2020:")
new_agebs_2020
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
Mex_merida_2020[Mex_merida_2020["CVEGEO"].isin(new_agebs_2020)].plot(ax=ax, color='blue', alpha=0.5, label='Nuevos AGEBs en 2020')
plt.title('New AGEBs 2020')
plt.legend()
plt.show()
```

## AGEBs that disappear from 2010 to 2020

```python
agebs_disappear = set_2010.difference(set_2020)
print("\nAGEBs that disappear from 2010 to 2020:")
agebs_disappear
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
Mex_merida_2010[Mex_merida_2010["CVEGEO"].isin(agebs_disappear)].plot(ax=ax, color='blue', alpha=0.5, label='AGEBs que desaparecen')
plt.title('AGEBs that disappear from 2010 to 2020')
plt.legend()
plt.show()
```

```python
# Create the plot
fig, ax = plt.subplots(figsize=(15, 15))

# Plot the AGEBs that remain from 2010 to 2020 (agebs_remain)
Mex_merida_2020[Mex_merida_2020['CVEGEO'].isin(agebs_remain)].plot(ax=ax, color='purple', edgecolor='black', label='Remain from 2010')

# Plot the new AGEBs in 2020 (new_agebs_2020)
Mex_merida_2020[Mex_merida_2020['CVEGEO'].isin(new_agebs_2020)].plot(ax=ax, color='blue', edgecolor='black', label='New in 2020')

# Plot the AGEBs that disappeared from 2010 to 2020 (disappeared_agebs)
Mex_merida_2010[~Mex_merida_2010['CVEGEO'].isin(agebs_remain)].plot(ax=ax, color='red', edgecolor='black', label='Disappeared by 2020')

# Adding legend and titles
plt.legend()
plt.title('AGEBs: Remain, New, and Disappeared')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
